```python
from fastapi import FastAPI
from pydantic import BaseModel

app = FastAPI()

class Question(BaseModel):
    question: str


@app.post("/ask")
def ask_ai(data: Question):

    response = llm.create_chat_completion(
        messages=[
            {"role": "system", "content": "You are Mentora AI Tutor."},
            {"role": "user", "content": data.question}
        ],
        max_tokens=200
    )

    answer = response["choices"][0]["message"]["content"]

    return {"answer": answer}
```

```python
# ============================================
# MENTORA AI Tutor - GGUF Version
# Model: TheBloke/Mistral-7B-Instruct-v0.2-GGUF
# Works in Google Colab
# ============================================

from llama_cpp import Llama

print("Libraries loaded")


# -------- 1. Model Selection --------
repo_id = "TheBloke/Mistral-7B-Instruct-v0.2-GGUF"

# Faster / lighter models you can switch to
model_file = "mistral-7b-instruct-v0.2.Q4_K_M.gguf"
# Alternatives:
# mistral-7b-instruct-v0.2.Q2_K.gguf (very small)
# mistral-7b-instruct-v0.2.Q5_K_M.gguf (better quality)

print("Loading model:", repo_id)


# -------- 2. Load Model --------
llm = Llama.from_pretrained(
    repo_id=repo_id,
    filename=model_file,
    n_ctx=4096,          # context length
    n_gpu_layers=35      # use GPU in Colab (T4)
)

print("Model loaded successfully!")


# -------- 3. Prompt Templates --------

def build_prompt(question):
    return f"""
You are MENTORA, a helpful AI Tutor.

Rules:
- Explain step by step
- Use simple language
- Give examples

Student Question: {question}

Tutor Answer:
"""


def hint_prompt(question):
    return f"""
You are MENTORA, an AI Tutor.

Give only a short hint.
Do NOT give the full answer.

Question: {question}

Hint:
"""


# -------- 4. Response Function --------

def generate_response(prompt, max_tokens=200):

    response = llm.create_chat_completion(
        messages=[
            {"role": "system", "content": "You are a helpful AI Tutor."},
            {"role": "user", "content": prompt}
        ],
        temperature=0.5,
        max_tokens=max_tokens
    )

    return response["choices"][0]["message"]["content"]


# -------- 5. Test Tutor Mode --------
print("\nTesting Tutor Mode")

question = "What is polymorphism in Java?"

print(generate_response(build_prompt(question)))


# -------- 6. Test Hint Mode --------
print("\nTesting Hint Mode")

question = "Write a Python program to reverse a string"

print(generate_response(hint_prompt(question), max_tokens=80))


# -------- 7. Interactive Mode --------
print("\nMENTORA Interactive Mode")

while True:

    mode = input("\nEnter mode (answer / hint / exit): ").lower()

    if mode == "exit":
        break

    question = input("Enter question: ")

    if mode == "answer":
        prompt = build_prompt(question)
        response = generate_response(prompt)

    elif mode == "hint":
        prompt = hint_prompt(question)
        response = generate_response(prompt, max_tokens=80)

    else:
        print("Invalid mode")
        continue

    print("\nMENTORA Response:\n")
    print(response)
```
